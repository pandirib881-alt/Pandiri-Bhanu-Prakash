```python

# Import necessary libraries
from surprise import Dataset, Reader, KNNBasic, accuracy
from surprise.model_selection import train_test_split

# Sample data: user-item interactions
# Format: user, item, rating
data_dict = {
    'user': ['User1', 'User1', 'User2', 'User2', 'User3', 'User3', 'User4', 'User4'],
    'item': ['ItemA', 'ItemB', 'ItemA', 'ItemC', 'ItemB', 'ItemC', 'ItemA', 'ItemB'],
    'rating': [5, 3, 4, 2, 4, 5, 3, 4]
}

# Convert to a pandas DataFrame
import pandas as pd
df = pd.DataFrame(data_dict)

# Define a Reader object with the rating scale
reader = Reader(rating_scale=(1, 5))

# Load data into Surprise dataset
data = Dataset.load_from_df(df[['user', 'item', 'rating']], reader)

# Split data into train and test sets
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

# Use User-Based Collaborative Filtering
algo = KNNBasic(sim_options={'name': 'cosine', 'user_based': True})

# Train the algorithm
algo.fit(trainset)

# Make predictions on the test set
predictions = algo.test(testset)

# Output the predictions
for pred in predictions:
    print(f'User: {pred.uid}, Item: {pred.iid}, Actual: {pred.r_ui}, Predicted: {pred.est:.2f}')

# Evaluate the overall performance
accuracy.rmse(predictions)
```
